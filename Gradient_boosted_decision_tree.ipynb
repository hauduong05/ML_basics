```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.ensemble import GradientBoostingClassifier
import matplotlib.pyplot as plt
import numpy as np
```

```python
bc = datasets.load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(bc.data, bc.target, random_state=0)
```

```python
gbdc = GradientBoostingClassifier(n_estimators=100, random_state=0, max_depth=2)
gbdc.fit(X_train, y_train)
```

```python
print(gbdc.score(X_train, y_train))
print(gbdc.score(X_test, y_test))
```

```python
y_pred = gbdc.predict(X_test)

print(classification_report(y_pred, y_test))
```

```python
num_feats = bc.data.shape[1]

fig = plt.figure(figsize=(16,10))
plt.barh(range(num_feats), gbdc.feature_importances_)
plt.yticks(np.arange(num_feats), bc.feature_names)
plt.xlabel('Feature Importance')
plt.ylabel('Feature')
plt.show()
```
